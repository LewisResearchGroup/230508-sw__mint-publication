## El-Maven

```python
maven = pd.read_csv('input/Maven-PeakAreaTop-Cleaned.csv')
df_maven = maven.set_index('label').filter(regex='2021').stack().reset_index().rename(columns={'level_1': 'ms_file_label', 0: 'maven', 'label': 'peak_label'})
df_maven
```

# Merge results

```python
merged = pd.merge(df_maven, mint.results, on=['ms_file_label', 'peak_label'])

merged = pd.merge(df_maven, mint.results, on=['ms_file_label', 'peak_label'])
merged = merged.rename(columns={'maven': 'Maven', 'peak_area_top3': 'ms-mint'})
                                                                           
#Symmetric percentage error                                                                           
merged['Î¼'] = 100 * ((merged['ms-mint'] - merged.Maven)/ (merged[['Maven', 'ms-mint']].max(axis=1)+1)).fillna(0)
merged['abs(Î¼)'] = merged['Î¼'].abs()

# Add metadata
merged = pd.merge(merged, meta, on='ms_file_label')
merged
```

```python
merged = merged.drop(['Unnamed: 0', 'plate_row', 'plate_col'], axis=1)
```

```python
merged = merged.drop(['concentration'], axis=1)
```

```python
merged['max(Maven, ms-mint)'] = merged[['ms-mint', 'Maven']].max(axis=1)
```
