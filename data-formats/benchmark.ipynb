# Get processing times

```python
import time
```

```python
n_files = 12

sets = [('mzXML', fn_mzXML), 
        ('mzML',  fn_mzML),
        ('mzMLb',  fn_mzMLb),
        ('parquet-TRR',  fn_parquet_trr),
        ('parquet-Mint',  fn_parquet_mint),
        ('feather-from-mzXML', fn_feather_from_mzXML),
        ('feather-from-mzML',  fn_feather_from_mzML), 
       ]                  

results = pd.DataFrame()
i = 0
for kind, fns in tqdm(sets):
    for fn in fns:
        for mode in ['Read only', 'Format']:
            print(fn)
            start = time.time()
            ms_file_to_df(fn, read_only = (mode=='Read only'))
            end = time.time()
            results.loc[i, ['File name', 'File format', 'Read time [s]', 'Mode']] = (fn, kind, end-start, mode)
            i += 1
```

```python
results['File size [MB]'] = results['File name'].apply(os.path.getsize)/1024/1024
results.to_csv(FN_RESULTS, index=False)
```
